```python
#Finding out the number of null values in all columns
sample1_gr11_11_39.isnull().sum()
```

**3.	Treat Missing Data in the Sample**

    **Approach followed**
    1. We first dropped the variables (columns) having less than 2/3 filled values 
    
    2. We then treated the missing data by filling the empty cells with means values and the mode depending on the variable

```python
# 1. Dropping columns with 2/3 (65814) empty entries
df1 = sample1_gr11_11_39.dropna(thresh=65814, axis =1)
df2 = pd.DataFrame(df1)
df2
```

```python
df2.describe()
```

```python
#2. Filling of the missing values
df2['Weekly_Sales'].fillna(float(df2['Weekly_Sales'].mean()), inplace=True)
df2
```

```python
df2['Dept'] = df2['Dept'].fillna(df2['Dept'].mode()[0])
df2
```

```python
df2['Date'] = df2['Date'].fillna(df2['Date'].mode()[0])
df2
```

```python
df2['IsHoliday'] = df2['IsHoliday'].fillna(df2['IsHoliday'].mode()[0])
df2
```

```python
#Result: all the missing data was treated successfully 
df2.isnull().sum()
```

**4.	Recode Features with Text Data into Numeric Data in the Sample (Do Not Delete the Features with Text Data. Create New Features with Equivalent Numeric Data). Use the Recoded Features for all Analysis**

```python
df2['Date'] = pd.Categorical(df2['Date']).codes
df2['IsHoliday'] = pd.Categorical(df2['IsHoliday']).codes
df2
```
